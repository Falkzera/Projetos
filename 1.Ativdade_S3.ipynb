#2 Exercício, treinando a conversão de um input em float

#3 Exercício, treinando um mini-sistema com if, elif e else

```python
name = input('Digite seu nome: ')
entry = input('Você quer entrar ou sair? ')

if entry == 'entrar':
    choice = f'Seja bem-vindo {name}, você entrou no sistema.'
elif entry == 'sair':
    choice = f'Até mais {name}, você saiu do sistema.'
else:
    choice = f'Opção inválida, {name}!'

print(choice)
```
